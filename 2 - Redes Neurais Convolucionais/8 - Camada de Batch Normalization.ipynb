No pré processamento, dados centralizados e normalizados aceleram a convergência do treinamento e fica mais fácil encontrar correlações.

Nesse contexto, foi proposta uma camada adicional nas CNNs após a convolução e antes da ativação não linear, a camada de normalização.

- Convolução -> **Batch Normalization** -> ReLU

A camada de normalização também possui pesos que são aprendidos, dessa forma podemos ter um tipo de normalização em uma convolução e outro tipo de normalização em outra convolução.

        batch_norm = nn.BatchNorm2d(qtd_features)

        Sendo qtd_features -> qtd de filtros da camada de convolução anterior.
